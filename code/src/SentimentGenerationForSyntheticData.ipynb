```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


# Initialize the analyzer
analyzer = SentimentIntensityAnalyzer()

    # Sample social media posts
posts = [
        "Kind of frustrated because thereâ€šÃ„Ã´s so much potential if they fix the bugs. Let me know if you've tried it too.",
        "Ugh, this update broke everything. So frustrating.",
        "It's okay, not bad but not amazing either.",
        "My package arrived early! Super happy with the service.",
        "Terrible experience. Never ordering from here again. ðŸ‘Ž"
    ]

for post in posts:
        score = analyzer.polarity_scores(post)
        compound = score['compound']
        sentiment = "Positive" if compound > 0.05 else "Negative" if compound < -0.05 else "Neutral"
        print(f"Post: {post}")
        print(f"Sentiment: {sentiment} | Score: {compound}")
        print("-" * 50)
```

```python
analyzer = SentimentIntensityAnalyzer()
df['sentiment'] = 'NA'

# Analyze each post
for index, row in df.iterrows():
        content = row['Content']
        score = analyzer.polarity_scores(content)
        compound = score['compound']
        sentiment = "Positive" if compound > 0.05 else "Negative" if compound < -0.05 else "Neutral"
       # df['Score'] = compound
       # df['sentiment'] = sentiment
        df.at[index, 'Content'] = content
        df.at[index, 'sentiment'] = sentiment
        df.at[index, 'Sentiment Score'] = compound
```

```python
# prompt: check if all the Customer_Id from df_sentiments are present either in Customer_id of df_cust or Customer_id of df_org

# Check if all Customer_Id from df_sentiments are present in df_cust or df_org
sentiments_customers = set(df_sentiments['Customer_Id'])
cust_customers = set(df_cust['Customer_Id'])
org_customers = set(df_org['Customer_Id'])

all_customers = cust_customers.union(org_customers)

if sentiments_customers.issubset(all_customers):
  print("All Customer_Id from df_sentiments are present in either df_cust or df_org.")
else:
  missing_customers = sentiments_customers.difference(all_customers)
  print(f"The following Customer_Id from df_sentiments are missing in both df_cust and df_org: {missing_customers}")
```
